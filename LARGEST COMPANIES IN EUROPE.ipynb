## WEB-SCRAPPING

### IMPORTING NECESSARY LIBRARIES

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_Europe_by_revenue'
```

### The purpose of this code to ask for permission for websites data,response other than[200] mean permission denied

```python
page = requests.get(url)
print(page)
```

### BeautifulSoup library is to parse HTML content and format it for easier readability,Prettify HTML content to make it more readable and visually appealing.

```python
soup = BeautifulSoup(page.text, 'html')
print(soup)
```

### The code is used to locate and extract a table (specifically its body) from an HTML document using BeautifulSoup. Here’s a detailed explanation of each part of the process:

```python
soup.find('table')
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
world_titles = table.find_all('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]

print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)

df
```

### The code is used to find all the rows in a table by searching for <tr> (table row) elements within a previously identified table or table body. Here’s a detailed explanation of the purpose of this line:

```python
column_data = table.find_all('tr')
```

### The code is used to extract REVENUE information from the rows of a table in a web page and store that information in a structured list. Here’s a detailed breakdown of each component and the overall purpose of the code:

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df.to_csv(r'Companies.csv', index = False)
```
